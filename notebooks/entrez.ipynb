```python
entrez = entrez.API(email="xyz@abc.com", name="rupav")
```

```python
entrez.databases
```

```python
search_results = entrez.search(db='gds', term=['HG-U133A', 'gpl'], usehistory='y')
```

##### self.time is a parameter just to keep track of API calls to NCBI through entrez utilities. Its s UNIX timestamp, initialized with 0 for the first request to NCBI.

#### Using 'usehistory' parameter above makes the server of NCBI to save the search results we just searched for, which can be used for other API endpoints in the form of "webenv" and "querykey" parameters

```python
search_results
WebEnv = search_results['webenv']
query_key = search_results['querykey']
(WebEnv, query_key)
```

#### So in next call to esummary endpoint, we don't need to provide "id" list as otherwise would have fetched from search_results.

```python
summary_results = entrez.summary(db='gds', id=None, WebEnv=WebEnv, query_key=query_key, retmax=20)
# these results after processing will be fed back to the user to choose accession number.
summary_results
```

```python
# following is one of the 20(see retmax parameter in above cell) dict values which will be sent to the user to choose from.
(summary_results[summary_results['uids'][0]])['title']
(summary_results[summary_results['uids'][0]])['summary']
(summary_results[summary_results['uids'][0]])['accession']
```

#### Now using above accession number we can download related CEL suppl files which contains gene expression data.

```python
accession = (summary_results[summary_results['uids'][0]])['accession']
search_results = entrez.search(db='gds', term=[accession, 'gse', 'cel'], usehistory='y', retmax=500)
WebEnv = search_results['webenv']
query_key = search_results['querykey']
```

```python
summary_results = entrez.summary(db='gds', id=None, WebEnv=WebEnv, query_key=query_key, retmax=20)
```

```python
links = []
series_accession_list = []
for key, value in summary_results.items():
    if key == 'uids':
        continue
    links.append(value.get('ftplink'))
    series_accession_list.append(value.get('accession'))
links
series_accession_list
```

```python
geo = GEO.API(path = '')  # path is where file will be downloaded.
# - not providing is equivalent to downloading at the current directory.
```

```python
geo.raw_data(links[0], series_accession_list[0])
```

### TAR file is downloaded in the above location, which can be then extracted to fetch CEL files.
