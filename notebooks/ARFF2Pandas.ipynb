```python
import pandas as pd
import arff as liacarff
```

```python
arff = liacarff.load(open('data/sample.arff', 'r'))
```

```python
arff
```

```python
attrs = arff['attributes']
data = arff['data']
```

```python
attrs_t = []
for attr in attrs:
    if isinstance(attr[1], list):
        attrs_t.append("{}@[{}]".format(attr[0], ', '.join(attr[1])))  # list for the possible values of the column
    else:
        attrs_t.append("{}@{}".format(attr[0], attr[1]))  # this indicates type of values/data-points in a column.
    
```

```python
attrs_t
```

```python
df = pd.DataFrame(data=arff['data'], columns=attrs_t)
```
